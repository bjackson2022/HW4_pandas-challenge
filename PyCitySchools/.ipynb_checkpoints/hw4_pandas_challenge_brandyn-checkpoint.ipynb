# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count =  school_data_complete['school_name'].nunique()
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete['Student ID'].nunique()
student_count
```

```python
# Calculate the total budget per school_create dataFrame
total_budget_per_school = school_data_complete.groupby('school_name')['budget'].mean()
total_budget_per_school.columns=['school_name','budget']

# Reset the index of the DataFrame
total_budget_per_school = total_budget_per_school.reset_index()

# Calculate the total budget
total_budget = total_budget_per_school['budget'].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate passing reading percentage
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["reading_score"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({'Total Schools': [school_count],'Total Students': [student_count],'Total Budget': [total_budget] ,'AVG Math Score':[average_math_score],'AVG Reading Score':[average_reading_score],'% Passing Math':[passing_math_percentage], '% Passing Reading':[passing_reading_percentage] ,'% Overall Passing Rate': [overall_passing_rate]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["% Overall Passing Rate"] = district_summary["% Overall Passing Rate"].round(2)
district_summary["% Passing Math"] = district_summary["% Passing Math"].round(2)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].round(2)
district_summary["AVG Math Score"] = district_summary["AVG Math Score"].round(2)
district_summary["AVG Reading Score"] = district_summary["AVG Reading Score"].round(2)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select the school type
school_types = school_data.set_index(["school_name"])["type"]
```

```python
# Calculate the total student count
per_school_counts = school_data_complete.groupby(["school_name"]).nunique()['Student ID']
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts
```

```python
# Calculate the average test scores
per_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
per_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
```

```python
# Filter the data to only include students who passed math
passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]

# Group by school_name and count the number of students who passed math
passing_math_students_by_school = passing_math.groupby("school_name").size()

#total students per school
total_students_per_school = school_data_complete.groupby("school_name").size()

#Find passing rate per school for students who passed math
total_pass_math__percent_per_school = passing_math_students_by_school / total_students_per_school
total_pass_math__percent_per_school = total_pass_math__percent_per_school.reset_index()
total_pass_math__percent_per_school = total_pass_math__percent_per_school.rename(columns={"school_name": "school_name", 0: "avg_pass_math"})

# Filter the data to only include schools that passed math with scores of 70 or higher
filter_total_schools_pass_math = total_pass_math__percent_per_school[(total_pass_math__percent_per_school["avg_pass_math"] >= 0.70)]

# Calculate the schools that passed math with scores of 70 or higher
total_schools_pass_math =  filter_total_schools_pass_math['school_name'].nunique()
total_schools_pass_math
```

```python
# Filter the data to only include students who passed reading
passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]

# Group by school_name and count the number of students who passed reading
passing_reading_students_by_school = passing_reading.groupby("school_name").size()

#total students per school
total_students_per_school = school_data_complete.groupby("school_name").size()

#Find passing rate per school for students who passed reading
total_pass_reading__percent_per_school = passing_reading_students_by_school / total_students_per_school
total_pass_reading__percent_per_school = total_pass_reading__percent_per_school.reset_index()
total_pass_reading__percent_per_school = total_pass_reading__percent_per_school.rename(columns={"school_name": "school_name", 0: "avg_pass_reading"})

# Filter the data to only include schools that passed reading with scores of 70 or higher
filter_total_schools_pass_reading = total_pass_reading__percent_per_school[(total_pass_reading__percent_per_school["avg_pass_reading"] >= 0.70)]

# Calculate the schools that passed reading with scores of 70 or higher
total_schools_pass_reading =  filter_total_schools_pass_reading['school_name'].nunique()
total_schools_pass_reading
```

```python
# Filter the data to only include students who passed both math and reading
passing_math_and_reading = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]

# Group by school_name and count the number of students who passed both math and reading
passing_math_and_reading_by_school = passing_math_and_reading.groupby("school_name").size()

# Group by school_name and count the number of students
total_math_sci_per_school = school_data_complete.groupby("school_name").size()

#Find passing rate per school for students who passed both math & science
total_math_sci_percent_per_school = passing_math_and_reading.groupby("school_name").size() / school_data_complete.groupby("school_name").size()

# Create DataFrame
total_math_sci_percent_per_school = passing_math_and_reading.groupby("school_name").size() / school_data_complete.groupby("school_name").size()
total_math_sci_percent_per_school = total_math_sci_percent_per_school.reset_index()
total_math_sci_percent_per_school = total_math_sci_percent_per_school.rename(columns={"school_name": "school_name", 0: "avg_pass_both_math_and_reading"})

# Filter the data to only include students with score >= 70
total_passing_math_sci_percent_per_school = total_math_sci_percent_per_school[(total_math_sci_percent_per_school["avg_pass_both_math_and_reading"] >= 0.70)]

# Calculate the schools that passed both math and reading with scores of 70 or higher
total_schools_passing_math_sci =  total_passing_math_sci_percent_per_school['school_name'].nunique()
total_schools_passing_math_sci
```

```python
# Use the provided code to calculate the passing math rates
per_school_passing_math = (school_data_complete[(school_data_complete["math_score"] >= 70)].groupby(["school_name"]).count()["Student ID"] / per_school_counts) * 100
per_school_passing_reading = (school_data_complete[(school_data_complete["reading_score"] >= 70)].groupby(["school_name"]).count()["Student ID"] / per_school_counts) * 100
overall_passing_rate = (school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby(["school_name"]).count()["Student ID"] / per_school_counts) * 100
```

```python
# Calculate the percent passing math, reading, and overall for each school
per_school_types = school_data_complete[['school_name', 'type']].drop_duplicates().set_index('school_name')['type']
per_school_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)].groupby("school_name").nunique()["Student ID"]
per_school_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)].groupby("school_name").nunique()["Student ID"]
per_school_passing_math_reading = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("school_name").nunique()["Student ID"]
per_school_counts = school_data_complete.groupby("school_name").nunique()["Student ID"]
per_school_math_scores = school_data_complete.groupby(["school_name"]).mean()["math_score"]
per_school_reading_scores = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts

percent_passing_math = per_school_passing_math / per_school_counts * 100
percent_passing_reading = per_school_passing_reading / per_school_counts * 100
percent_overall_passing = per_school_passing_math_reading / per_school_counts * 100

# Create a dataframe to hold the above results
per_school_summary = pd.DataFrame({
    "School Type": per_school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": per_school_math_scores,
    "Average Reading Score": per_school_reading_scores,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_reading,
    "% Overall Passing": percent_overall_passing
})

# Formatting
per_school_summary['Total School Budget'] = per_school_summary['Total School Budget'].map('${:,.0f}'.format)
per_school_summary['Total Students'] = per_school_summary['Total Students'].map('{:,.0f}'.format)
per_school_summary['Per Student Budget'] = per_school_summary['Per Student Budget'].map('${:,.2f}'.format)
per_school_summary['Average Math Score'] = per_school_summary['Average Math Score'].map('{:.2f}'.format)
per_school_summary['Average Reading Score'] = per_school_summary['Average Reading Score'].map('{:.2f}'.format)
per_school_summary['% Passing Math'] = per_school_summary['% Passing Math'].map('{:.1f}%'.format)
per_school_summary['% Passing Reading'] = per_school_summary['% Passing Reading'].map('{:.1f}%'.format)
per_school_summary['% Overall Passing'] = per_school_summary['% Overall Passing'].map('{:.1f}%'.format)

# Display the DataFrame
per_school_summary 
```
